```python
# @title Importing Pandas and making a dataframe from the dataset using Pandas

import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/creditcard.csv")
df
```

```python
# @title Finding Null Values for every column; no null values found
df.isnull().sum()
```

```python
# @title Description of every column
df.describe()
```

```python
# @title Scaling the data using Standard Scaler
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
scaled_df=pd.DataFrame(ss.fit_transform(df))
```

```python
# @title Looking for correaltion among columns using Seaborn and MatPlotLib; "Time" and "V1" to "V8", "Amount" and "Class" are correlated.
import seaborn as sns
import matplotlib.pyplot as plt
corr_matrix = df.corr(method="spearman")
sns.heatmap(corr_matrix > 0.8, annot=True, cbar=False)
plt.show()
```

```python
# @title The output column "Class" is highly imbalanced as seen.
df['Class'].value_counts()
```

```python
# @title Using Random Under Sampling to fix the imbalance in dataset.
from imblearn.under_sampling import RandomUnderSampler
x=df[['Time','V1','V2','V3','V4','V5','V6','V7','V8','V9','V10','V11','V12','V13','V14','V15','V16','V17','V18','V19','V20','V21','V22','V23','V24','V25','V26','V27','V28','Amount']]
y=df['Class']
rus=RandomUnderSampler()
rus_x,rus_y=rus.fit_resample(x,y)
rus_y.value_counts()
```

```python
# @title Performing the Train Test Split on Dataset
input_data=rus_x
output_data=rus_y
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input_data,output_data,test_size=0.25)
```

```python
# @title Using Pruning Technique to find optimal number of depth for Decision Tree
from sklearn.tree import DecisionTreeClassifier
for i in range(1,20):
  dt=DecisionTreeClassifier(max_depth=i)
  dt.fit(x_train,y_train)
  print(dt.score(x_train,y_train)*100,dt.score(x_test,y_test)*100,i)
```

```python
# @title Using Decision Tree Classifier and plotting it
dt1=DecisionTreeClassifier(max_depth=2)
dt1.fit(x_train,y_train)
from sklearn.tree import plot_tree
plot_tree(dt1)
plt.show()
```

```python
# @title Using Support Vector Machine for classification
from sklearn.svm import SVC
sv=SVC(kernel='sigmoid')
sv.fit(x_train,y_train)
```

```python
# @title Using Bagging Classifier (Ensemble Technique)with K Nearest Neighbour
from sklearn.ensemble import BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
bg=BaggingClassifier(estimator=KNeighborsClassifier())
bg.fit(x_train,y_train)
```

```python
# @title Using Random Forest Classification
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
# @title Comparing the scores of all techniques to get best one.
print("score of decision tree",dt1.score(x_train,y_train)*100-dt1.score(x_test,y_test)*100)
print("score of support vector machine",sv.score(x_train,y_train)*100-sv.score(x_test,y_test)*100)
print("score of bagging classifier",bg.score(x_train,y_train)*100-bg.score(x_test,y_test)*100)
print("score of random forest",rf.score(x_train,y_train)*100-rf.score(x_test,y_test)*100)
print("best technique is decision tree")
```
